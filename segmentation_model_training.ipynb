```python
x_train = np.load('/content/drive/MyDrive/Semantic_Segmentation/x_train.npy')
y_train = np.load('/content/drive/MyDrive/Semantic_Segmentation/y_train.npy')
x_val = np.load('/content/drive/MyDrive/Semantic_Segmentation/x_val.npy')
y_val = np.load('/content/drive/MyDrive/Semantic_Segmentation/y_val.npy')
```

```python
import segmentation_models as sm
# from keras.layers import Input, Conv2D, Lambda
# from keras.models import Model
```

```python
# https://stackoverflow.com/questions/67792138/attributeerror-module-keras-utils-has-no-attribute-get-file-using-segmentat
sm.set_framework('tf.keras')
```

```python
BACKBONE = 'resnet34'
preprocess_input = sm.get_preprocessing(BACKBONE)
```

```python
x_train = preprocess_input(x_train)
x_val = preprocess_input(x_val)
```

```python
N = x_train.shape[-1] # -> 3
n_classes = y_train.shape[-1] # -> 6

model = sm.Unet(backbone_name='resnet34', classes=n_classes, activation='softmax', encoder_weights='imagenet', encoder_freeze=True)
```

```python
# model.summary()
```

```python
model.compile(optimizer='Adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x=x_train, y=y_train, batch_size=16, epochs=100, validation_data=(x_val, y_val))
```

```python
model.save('/content/drive/MyDrive/Semantic_Segmentation')
```
